## Linear regression

```python
#Linear regression

import numpy as np
import matplotlib.pyplot as plt

x= np.random.normal(3,1,1000)
y= 100-(x+np.random.normal(0,0.1,1000))*3

plt.scatter(x,y)

def predit(x):
    return -3*x+100

fitline=predit(x)
plt.plot(x, fitline,c='b')
```

## Polymonial regression

```python
#Polymonial regression
%matplotlib inline
from pylab import *
import numpy as np

np.random.seed(4)
x=np.random.normal(3.0,1.0,1000)
y=np.random.normal(50.0,10.0,1000)/x

plt.scatter(x,y)
```

```python
a=np.array(x)
b=np.array(y)

p9= np.poly1d(np.polyfit(a,b,9))

xp=np.linspace(0,6,3000)
plt.scatter(x,y)
plt.plot(xp,p9(xp),c='b')
```
